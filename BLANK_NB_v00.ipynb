## Import Dependencies

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## Set Variables

```python
path = './KaggleCache/datasets/andrewmvd/ocular-disease-recognition-odir5k/versions/2'
df = pd.read_csv(os.path.join(path, 'full_df.csv'))
```

## Example MD Cell #1
Shows the first 5 entries of the pandas dataframe

```python
# Display the header
# Set options for a compact view
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', 1000)        # Prevent wrapping
pd.set_option('display.max_colwidth', 15)   # Truncate long text (like file paths)

df.head(3)
display(df.head())
```

## Example MD Cell #2
Shows the image that correlates to the first entry in the dataframe

```python
# 1. Get a filename from the dataframe (e.g., the first row)
sample_filename = df['filename'][0]

# 2. Construct the path specifically pointing to the 'preprocessed_images' folder
# Structure: path -> preprocessed_images -> filename
image_path = os.path.join(path, 'preprocessed_images', sample_filename)

# 3. Load and display the image directly
img = mpimg.imread(image_path)
plt.imshow(img)
plt.title(f"Filename: {sample_filename}")
plt.axis('off')
plt.show()
```
