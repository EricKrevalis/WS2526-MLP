# Run this Cell to delete KaggleCache folder

```python
import shutil
import os

# --- CONFIGURATION ---
TARGET_FOLDER = 'KaggleCache'
# ---------------------

print(f"Scanning for folder '{TARGET_FOLDER}' starting from: {os.getcwd()} ...")
print("-" * 40)

found_and_deleted = False

# 1. Walk the directory tree (Top-down)
for root, dirs, files in os.walk(os.getcwd()):
    if TARGET_FOLDER in dirs:
        # 2. Construct the Absolute Path (Bypassing relative path issues)
        absolute_path = os.path.join(root, TARGET_FOLDER)
        
        print(f"Found target at: {absolute_path}")
        
        # 3. Delete it
        try:
            shutil.rmtree(absolute_path)
            print("DELETED successfully.")
            found_and_deleted = True
            
            # Optional: Remove this if you want to delete MULTIPLE copies in subfolders
            # breaking here ensures we only kill the first/main one we find.
            break 
            
        except OSError as e:
            print(f"Error deleting folder: {e.strerror}")

if not found_and_deleted:
    print(f"Could not find '{TARGET_FOLDER}' anywhere inside the current directory tree.")
```
